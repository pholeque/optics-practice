```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from scipy import optimize
```

```python
file = open('malus.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
x = data[:, 0]
y = data[:, 1]
t = np.linspace(-25, 335, 360)
plt.figure(figsize=(12, 7))
plt.plot(t+25, 500*(np.cos(t*np.pi/180))**2, label="$I_0cos^2(alpha)$")
plt.errorbar(x, y,0.05 *y, 2.5,  color='red', fmt='.', label="data")
plt.xlabel("$alpha, ^o$")
plt.ylabel("Intensity")


plt.grid() 
plt.title("Malus's law")
plt.legend()
plt.savefig("malus.pdf")
plt.show()
```

```python
file = open('2_9.txt', 'r')
file1 = open('1_9.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
x = data[:, 0]
y = data[:, 1]/180

plt.figure(figsize=(10, 8.5))
t = np.linspace(15, 75, 600)
data1 = np.genfromtxt(file, delimiter=',')
x1 = data[:, 0]
y1 = data[:, 1]/500

r01 = ((np.cos((t+1)*np.pi/180)-np.sqrt(1.77-(np.sin((t+1)*np.pi/180))**2))/(np.cos((t+1)*np.pi/180)+np.sqrt(1.77-(np.sin((t+1)*np.pi/180))**2)))**2
r00 = ((np.cos((t-1)*np.pi/180)-np.sqrt(1.5-(np.sin((t-1)*np.pi/180))**2))/(np.cos((t-1)*np.pi/180)+np.sqrt(1.5-(np.sin((t-1)*np.pi/180))**2)))**2
r = ((np.cos(t*np.pi/180)-np.sqrt(1.64-(np.sin(t*np.pi/180))**2))/(np.cos(t*np.pi/180)+np.sqrt(1.64-(np.sin(t*np.pi/180))**2)))**2
plt.plot(t, r, color='red', label="s-polarization")
plt.plot(t, r00, color='red', linestyle ='--')
plt.plot(t, r01, color='red', linestyle ='--')
plt.errorbar(x, y,   color='blue', fmt='.')

r11 = ((-(1.28-0.05)*np.cos((t-1)*np.pi/180)+np.sqrt(1-(np.sin((t-1)*np.pi/180)/(1.28-0.05))**2))/((1.28-0.05)*np.cos((t-1)*np.pi/180)+np.sqrt(1-(np.sin((t-1)*np.pi/180)/(1.28-0.05))**2)))**2
r10 = ((-(1.28+0.05)*np.cos((t+1)*np.pi/180)+np.sqrt(1-(np.sin((t+1)*np.pi/180)/(1.28+0.05))**2))/((1.28+0.05)*np.cos((t+1)*np.pi/180)+np.sqrt(1-(np.sin((t+1)*np.pi/180)/(1.28+0.05))**2)))**2
r1 = ((-1.28*np.cos(t*np.pi/180)+np.sqrt(1-(np.sin(t*np.pi/180)/1.28)**2))/(1.28*np.cos(t*np.pi/180)+np.sqrt(1-(np.sin(t*np.pi/180)/1.28)**2)))**2
plt.plot(t, r1, color='blue', label="p-polarization")
plt.plot(t, r10, color='blue', linestyle ='--')
plt.plot(t, r11, color='blue', linestyle ='--')

plt.errorbar(x1, y1,  color='green', fmt='.')
plt.xlabel("$alpha, ^o$")
plt.ylabel("R")
plt.vlines(52, -0.05, 0.3,
          color = 'black',
          label = "Bruster's angle",
          linewidth = 1.2,
          linestyle = '--')

plt.ylim(-0.05, 0.3)

plt.grid() 
plt.title("Fresnel's Equations")
plt.legend()
plt.savefig("frenel.pdf")
plt.show()
```
