```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from scipy import optimize
```

```python
file = open('3_0.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
sint = (data[:, 0]-188.5)/np.sqrt((data[:, 0]-188.5) ** 2 + 122 ** 2)
x = data[:, 1]
plt.figure(figsize=(10, 7))
plt.errorbar(x, sint, color='red', fmt='.', label="data")
plt.xlabel("m")
plt.ylabel("sin (theta)")
a0, b = np.polyfit(x, sint, deg=1)
plt.plot(x, a0 * x + b, 'b', label='fit: 0.1558x, n = 299.54 $mm^{-1}$')

tab = np.vstack((np.around(data[:, 1]), np.around(data[:, 0], 1), np.around(sint, 2))).T
print(tab)
print(1/(520/a0)*10 ** 6)
plt.grid() 
plt.title("sin (theta)")
plt.legend()
#plt.savefig("3_0.pdf")
plt.show()
```

```python
file = open('1_2.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
t = (data[:, 0]-100)/np.sqrt((data[:, 0]-100) ** 2 + 119.5 ** 2) 
y = t - np.sin(0.26)+0.5
x = data[:, 1]
plt.figure(figsize=(10, 7))
plt.errorbar(x, y, color='red', fmt='.', label="data")
plt.xlabel("m")
plt.ylabel("sin (theta)")
a, b = np.polyfit(x, y, deg=1)
plt.plot(x, a * x + b, 'g', label='fit: 0.309.6x, $n = 595 mm^{-1}$')


k = y / x
k = np.around(1/(520/np.delete(k, 3)) * 10 ** 6,2) - np.around(1/(520/a) * 10 ** 6,2)
print(k)
print("коэф-т наклона =", a)
print('число штрихов с учетом среднеквадр. отклонения',np.around(1/(520/a) * 10 ** 6,2), '+-', np.around(np.sqrt(np.sum(np.power(k, 2))/8),2)) 


plt.grid() 
plt.title("sin (theta)")
plt.legend()
plt.savefig("1_2.pdf")
plt.show()
```
