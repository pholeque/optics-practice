```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from scipy import optimize
import scipy as sp
```

```python
file = open('X_red_1.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
x = np.sin(data[:, 0]*np.pi/180) 
y = np.sin(data[:, 1]*np.pi/180) 
plt.plot(x, y,'.', label='experimental data', color = 'red')
#plt.plot(x, a0 * x + b, 'b', label='linear approximation')
t = np.linspace(0.45,0.8,100)
def test_func(x, a, b):
    return np.sqrt(b ** 2 - (500/a*x)**2)
params, params_covariance = optimize.curve_fit(test_func, x, y, p0=[215, 2])
print(params)
plt.plot(x, test_func(x, params[0], params[1]), label=r'$(1.38^2 - (500/260*sin\phi)^2)^{0.5}$')


plt.grid()
plt.xlabel(r'sin$\phi$')
plt.ylabel(r'sin$\theta$')
plt.title(r'sin$\theta(sin\phi) - X$')
plt.legend()
plt.savefig("X_red_1_sin.pdf")
plt.show()
```

```python
file = open('X_red_1.txt', 'r')
data = np.genfromtxt(file, delimiter=',')
x = np.sin(data[:, 0]*np.pi/180) ** 2
y = np.sin(data[:, 1]*np.pi/180) ** 2
plt.errorbar(x, y, 0.04,0,'.', label='experimental data', color = 'red')
def test_func(x, n, d):
    return n ** 2 - (500/d)**2 * x
params, params_covariance = optimize.curve_fit(test_func, x, y, p0=[215, 2])

plt.plot(x, test_func(x, params[0], params[1]))

print('n =',params[0], 'D =',params[1])
print(np.sum(abs(y - test_func(x, params[0], params[1])))/len(x))
plt.grid()
plt.xlabel(r'sin$^2\phi$')
plt.ylabel(r'sin$^2\theta$')
plt.title(r'sin$^2\theta(sin^2\phi) - X$')
plt.legend()
plt.savefig("X_red_1_sin.pdf")
plt.show()
```
